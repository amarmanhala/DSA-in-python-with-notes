```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None
    self.len = 0

  def __str__(self):
    current = self.head
    result = ''
    while current != None:
      result = result + str(current.data) + ' '
      current = current.next
    return result[:-2]


  def insert_at_head(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node
    self.len = self.len + 1

  def append(self, data):

    new_node = Node(data)

    if self.head == None:
      self.head = new_node
      self.len = self.len + 1

    else:

      current = self.head
      while current.next != None:
        current = current.next
      current.next = new_node
      self.len = self.len + 1

  def insert_after_at_given_item(self, data, item):
    new_node = Node(data)
    current = self.head
    while current != None:
      if current.data == item:
       break
      current = current.next
    if current != None:
      new_node.next = current.next
      current.next = new_node
      self.len = self.len + 1

    else:
      return "Item not found"

  def clear(self):
    self.head = None
    self.len = 0

  def delete_head(self):
    if self.head == None:
      return "List is empty"
    else:
      self.head = self.head.next
      self.len = self.len - 1

  def delete_tail(self):
    if self.head == None:
      return "List is empty"
    else:
      current = self.head
      if current.next == None:
        return self.delete_head()
      while current.next.next != None:
        current = current.next
      current.next = None
      self.len = self.len - 1

  def remove(self, item):

    if self.head.data == item:
      return self.delete_head()

    current = self.head

    while current.next != None:
      if current.next.data == item:
        break
      current = current.next
    current.next = current.next.next

  def search(self, item):

    if self.head == None:
      return "List is empty"

    curr = self.head
    index = 0

    while curr != None:
      if curr.data == item:
        return f"Item {item} found at index {index}"

      curr = curr.next
      index += 1
    return "Item not found"

  def __getitem__(self, index):

    if self.head == None:
      return "List is empty"

    curr = self.head
    pos = 0
    while curr != None:
      if pos == index:
        return curr.data
      curr = curr.next
      pos += 1

    return "Item not found!"

  def replace_given_item_with_maximum_value(self, item):
    curr = self.head
    max = curr
    while curr.next != None:
      if curr.data > max.data:
        max = curr
      curr = curr.next
    max.data = item
    return True


  def sum_of_values_on_odd_position(self):
    curr = self.head
    pos = 0
    sum = 0

    while curr != None:
      if pos % 2 != 0:
        sum += curr.data
      curr = curr.next
      pos += 1
    return sum

  def reverse(self):
    curr_node = self.head
    prev = None

    while curr_node != None:
      next_node = curr_node.next
      curr_node.next = prev
      prev = curr_node
      curr_node = next_node

    self.head = prev

  def string_pattern(self):
    curr = self.head

    while curr != None:

      if curr.data == '/' or curr.data == '*':
        curr.data = ' '

        if curr.next.data == '/' or curr.next.data == '*':
          curr.next.next.data = curr.next.next.data.upper()
          curr.next = curr.next.next


      curr = curr.next





```
